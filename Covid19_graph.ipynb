```python
import pandas as pd
import matplotlib.pyplot as plt
import pymongo 
import numpy as np
import matplotlib
client = pymongo.MongoClient("mongodb+srv://siphe:siphesihle@reportingdata.14tbm.mongodb.net/WHO?retryWrites=true&w=majority")
# Database Name
db = client["WHO"]

# Collection Name
col = db["Covid"]

covid = col.find()

df = pd.DataFrame(covid)

gp = df[['TotalDeath', 'TotalCase', 'NewDeath', 'NewCase', 'COUNTRY_NAME', 'Date_epicrv']].groupby(['Date_epicrv'], as_index=False).sum()

```

```python
gp = df[['TotalDeath', 'TotalCase', 'NewDeath', 'NewCase', 'COUNTRY_NAME']].groupby(['COUNTRY_NAME'], as_index=False).sum()

plt.figure(figsize=(25,8))
plt.bar(gp['COUNTRY_NAME'].head(15), gp['NewCase'].head(15))
plt.xticks(rotation=90)
plt.title('New cases')
plt.show()
```

```python
gp = df[['TotalDeath', 'TotalCase', 'NewDeath', 'NewCase', 'COUNTRY_NAME']].groupby(['COUNTRY_NAME'], as_index=False).sum()

plt.figure(figsize=(25,10))
plt.scatter(gp['COUNTRY_NAME'].head(15), gp['NewCase'].head(15))
plt.xticks(rotation=90)
plt.title('Tota')
plt.show()
```
